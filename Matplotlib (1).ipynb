```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Read data from CSV file
data = pd.read_csv('F:\All_Race_Numbers.csv')

# Read columns
years = data['Year'].tolist()
black_values = data['Black'].tolist()

# Create bar graph
plt.figure(figsize=(10, 6))
bars = plt.bar(years, black_values, color='black')

# Customizing y-axis ticks
def millions_formatter(x, pos):
    'The two args are the value and tick position'
    if x >= 1000000:
        return f'{x / 1000000:.0f}M'
    else:
        return f'{x / 1000:.0f}K'

formatter = FuncFormatter(millions_formatter)
plt.gca().yaxis.set_major_formatter(formatter)

# Setting y axis limit
max_value = max(black_values)
plt.ylim(0, max_value * 1.1)  

# Add title and labels
plt.title('Black Column Values Over Years')
plt.xlabel('Years')
plt.ylabel('Black')

# Rotate x axis labels if needed
plt.xticks(rotation=45)

# Label each bar with its value
for bar, value in zip(bars, black_values):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 100000, f'{value:,}', ha='center', va='bottom')

# Show plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from CSV file
data = pd.read_csv('F:\All_Race_Numbers.csv')

# Filter data for the year 2019
data_2019 = data[data['Year'] == 2019]

# Read columns
columns = ['White', 'Black', 'Hispanic', 'Asain/Native Hawaiian and Pacific Islander', 'American Indian/Alaska Native', 'Multiple Races']

# Get the sum of each column for the year 2019
values = [data_2019[column].sum() for column in columns]

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(values, labels=columns, autopct='%1.1f%%', startangle=140)

# Add title
plt.title('Percentages of Race in Poverty in 2019')

# Show plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Read the data from the CSV file
data = pd.read_csv('F:\All_Race_Numbers.csv')

# Assuming the CSV file has columns 'Year', 'Total in Poverty', and 'Black in Poverty'
years = data['Year']
total_in_poverty = data['Total']
black_in_poverty = data['Black']

# Plotting the total number of people in poverty and filling the area under the line
plt.plot(years, total_in_poverty, color='lightblue', label='Total in Poverty', marker='o')
plt.fill_between(years, total_in_poverty, color='lightblue', alpha=0.3)

# Plotting the number of black people in poverty and filling the area under the line
plt.plot(years, black_in_poverty, color='orange', label='Black in Poverty', marker='o')
plt.fill_between(years, black_in_poverty, color='orange', alpha=0.3)

# Setting y-axis limit to 50,000,000
plt.ylim(0, 50000000)

# Customizing y-axis ticks
def millions_formatter(x, pos):
    'The two args are the value and tick position'
    if x >= 1000000:
        return f'{x / 1000000:.0f}M'
    else:
        return f'{x / 1000:.0f}K'

formatter = FuncFormatter(millions_formatter)
plt.gca().yaxis.set_major_formatter(formatter)

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Number of People')
plt.title('Total Number of People in Poverty with Black Population Highlighted')

# Displaying the legend
plt.legend()

# Displaying the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from CSV file
data = pd.read_csv('F:\All_Race_Percentages.csv')

# Define racial groups
racial_groups = ['Black', 'White', 'Asain/Native Hawaiian and Pacific Islander', 'Hispanic', 'American Indian/Alaska Native', 'Multiple Races']

# Create subplots for each racial group
fig, axs = plt.subplots(len(racial_groups), figsize=(10, 8), sharex=True, sharey=True)

# Iterate over each racial group and create a histogram
for i, racial_group in enumerate(racial_groups):
    group_data = data[racial_group].tolist()
    axs[i].hist(group_data, bins=10, color='skyblue', edgecolor='black')
    axs[i].set_title(racial_group)
    axs[i].set_xlabel('Poverty Rate')  # Set x-axis label for each subplot
    axs[i].yaxis.set_major_locator(plt.MaxNLocator(3))  # Set maximum of 3 ticks on y-axis

# Add common y axis label and title
fig.text(0.04, 0.5, 'Frequency', va='center', rotation='vertical')
plt.suptitle('Distribution of Poverty Rates Across Racial Groups')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from CSV file
data = pd.read_csv('F:\All_Race_Percentages.csv')

# Extract relevant columns for Black individuals
black_poverty_rate = data['Black']
black_dropout_rate = data['Black DO']
years = data['Year']

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(black_poverty_rate, black_dropout_rate, c=years, cmap='viridis', edgecolor='black', alpha=0.8)

# Add diagonal line representing perfect positive correlation
plt.plot([0, max(black_poverty_rate)], [0, max(black_dropout_rate)], color='red', linestyle='--', label='Perfect Correlation')

# Add labels, title, and legend
plt.xlabel('Black Poverty Rate')
plt.ylabel('Black Dropout Rate')
plt.title('Relationship Between Black Poverty Rate and Dropout Rate')
plt.legend()

# Add colorbar to show years
cbar = plt.colorbar()
cbar.set_label('Year')

# Show plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
